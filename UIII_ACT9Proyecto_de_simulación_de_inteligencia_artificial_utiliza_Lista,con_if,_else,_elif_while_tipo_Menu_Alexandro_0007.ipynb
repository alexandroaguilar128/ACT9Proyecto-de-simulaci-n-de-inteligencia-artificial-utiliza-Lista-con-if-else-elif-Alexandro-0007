### Proyecto de simulación de inteligencia artificial  utiliza Lista,con if, else, elif  while tipo Menu.
### Medico

```python
atendiendo = True
pacientes = []
conteo = 0

print("--- SISTEMA MÉDICO IA: CONSULTORIO VIRTUAL ---")

# Ciclo principal de interacción [5.2, 1243]
while atendiendo == True:
    print("\nMENÚ DE OPCIONES:")
    print("1. Registrar paciente")
    print("2. Mostrar lista de espera")
    print("3. Realizar diagnóstico lógico")
    print("4. Finalizar sistema")
    opc = input("Seleccione una opción (1-4): ")

    # CASO 1: Registro (Uso de if inicial) [4.3, 1242]
    if opc == "1":
        nombre = input("IA: Nombre del paciente: ")
        pacientes.append(nombre)
        conteo = conteo + 1
        print("IA: Registro completado.")

    # CASO 2: Listado manual (Uso de primer elif necesario) [5.2, 1243]
    elif opc == "2":
        if conteo == 0:
            print("IA: No hay pacientes en la lista.")
        else:
            indice = 0
            while indice < conteo:
                print(str(indice + 1) + ". " + pacientes[indice])
                indice = indice + 1

    # CASO 3: Diagnóstico IA (Uso de segundo elif necesario) [4.6, 1242]
    elif opc == "3":
        if conteo == 0:
            print("IA AVISO: Debe registrar a alguien primero.")
        else:
            sintoma = input("IA: ¿Cuál es el síntoma? (tos / dolor / ninguno): ")
            temp = float(input("IA: ¿Cuál es la temperatura? (°C): "))
            if temp > 38.0 and sintoma == "tos":
                print("IA DIAGNÓSTICO: Infección respiratoria probable.")
            elif temp > 37.0 and sintoma == "dolor":
                print("IA DIAGNÓSTICO: Posible cuadro inflamatorio.")
            else:
                print("IA DIAGNÓSTICO: Estado estable o no determinado.")

    # CASO 4: Salida (Uso de tercer elif necesario) [5.2]
    elif opc == "4":
        print("IA: Apagando sistema... ¡Salud para todos!")
        atendiendo = False
print("Elaborado por Alexandro Aguilar NL = 1")
```
